```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import riders/riders-normalized.arrow
path = './riders/riders-normalized.csv'
df = pd.read_csv(path)
df.columns
```

```python
filtered_df = df[df['year'] == 2014]
filtered_df.head()
```

```python
# create a function that takes a df and year, and returns the correlation coefficient between age and rank
def age_rank_corr(df, year):
    df_year = df[df['year'] == year]
    return df_year['rider.age'].corr(df_year['rank'])

# loop through all years from 2013 to 2023 and print the correlation coefficient

for year in range(2013, 2024):
    print(f'Correlation coefficient for {year}: {age_rank_corr(df, year)}')

# plot the correlation coefficient for each year
plt.plot(range(2013, 2024), [age_rank_corr(df, year) for year in range(2013, 2024)])
    
```
